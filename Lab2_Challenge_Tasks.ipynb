```python
#Redaction Protocol Implementation

# Initialize variables as specified
target_name = "MAÑIBO"  # Original Surname (uppercase)
SEED_NUM = 2            # Last digit of STUDENT_ID
redacted_name = ""

# Define vowels (including Ñ? Note: Per problem statement, vowels are A,E,I,O,U; Ñ is consonant here)
VOWELS = {'A', 'E', 'I', 'O', 'U'}

# Iterate through each character to process
for char in target_name:
    if char in VOWELS:
        redacted_name += str(SEED_NUM)
    else:
        redacted_name += char

# Execution Log and Output
print("Assessment Data:")
print(f"Original Surname: {target_name}")
print(f"SEED_NUM Used: {SEED_NUM}")
print("Execution Log: Redaction process completed successfully")
print(f"Redacted Output: {redacted_name}")
```
